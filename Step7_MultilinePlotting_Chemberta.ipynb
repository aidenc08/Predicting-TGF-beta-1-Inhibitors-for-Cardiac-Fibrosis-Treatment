```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.size'] = 10
plt.rcParams['svg.fonttype'] = 'none'
```

```python
import pandas as pd
path = "/content/drive/Shareddrives/1:1_Aiden_Chavda/Results/Accuracy Metric Comparison - Chemberta_R2.csv"

data = pd.read_csv(path)
data.head()
```

```python
epochs = list(data["Epochs/LR"])
lr = list(data.columns)
lr = lr[1:]
all_score = []
for index in range(len(epochs)):
  all_score.append(list(data.iloc[index,1:]))

fig, ax = plt.subplots(1,1,figsize=(5,5))
for index in range(len(epochs)):
  ax.plot(lr, all_score[index], label = "Epochs: {}".format(epochs[index]))
ax.set_xlabel("Learning Rate")
ax.set_ylabel("R2 Score")
ax.set_title("R2 Score vs. Learning Rate and Epochs", size = 10)
ax.grid(True)
ax.legend(loc = "best")
plt.tight_layout()
plt.savefig("/content/drive/Shareddrives/1:1_Aiden_Chavda/Results/Chemberta_Multiline_Plot.png")
```

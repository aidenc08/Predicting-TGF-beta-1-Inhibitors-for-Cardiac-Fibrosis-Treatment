```python
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.size'] = 10
plt.rcParams['svg.fonttype'] = 'none'
```

```python
import pandas as pd
path = "/content/drive/Shareddrives/1:1_Aiden_Chavda/Results/Accuracy Metric Comparison - RandomForest_R2.csv"

data = pd.read_csv(path)
data.head()
```

```python
df = []
for i in range(10):
  temp = []
  for j in range(7):
    temp.append(data['Mean r2 score'][i*7 + j])
  df.append(temp)

df = pd.DataFrame(data = df, columns = [str(i) for i in range(1, 8)], index = [str(i*10) for i in range(1, 11)])
df = df.transpose()
df
```

```python
estimators = list(df.columns)
max_depth = list(df.index)
all_score = []
for index in range(len(max_depth)):
  all_score.append(list(df.iloc[index,0:]))

plt.figure(figsize = (4,4))
for index in range(len(max_depth)):
  plt.plot(estimators, all_score[index], label = "Max Depth: {}".format(max_depth[index]))
plt.xlabel("Number of Estimators")
plt.ylabel("R2 Score")
plt.title("R2 Score with Estimators and Max Depth", size = 10)
plt.grid(True)
plt.legend(loc = "best")
plt.show()
```

```python
#df = df.transpose()
estimators = list(df.index)
max_depth = list(df.columns)
all_score = []
for index in range(len(estimators)):
  all_score.append(list(df.iloc[index,0:]))

fig, ax = plt.subplots(1,1,figsize=(4,4))
for index in range(len(estimators)):
  ax.plot(max_depth, all_score[index], label = "Estimators: {}".format(estimators[index]))
ax.set_xlabel("Max Depth")
ax.set_ylabel("R2 Score")
ax.set_title("R2 Score vs. Estimators and Max Depth", size = 10)
ax.grid(True)
ax.legend(loc = "best")
plt.tight_layout()
plt.savefig("/content/drive/Shareddrives/1:1_Aiden_Chavda/Results/Random_Forest_Multiline_Plot.png")
```
